```python
import numpy as np
import h5py
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
path='./Results/'

columns = ['noise_level', 'n_train', 'dt', 'eq', 'mse_train', 'mse_val', 'phase_noise', 'r2_score_train',
            'r2_score_val', 'threshold_values']
df = pd.DataFrame(columns=columns, index=range(len(os.listdir(path))))

for j,filename in enumerate(os.listdir(path)):
    noise_level = float(filename.split('noise_')[1].split('_')[0])
    n_train = int(filename.split('train_')[1].split('_')[0])
    dt = float(filename.split('dt_')[1].split('.h')[0])
    
    file = h5py.File(os.path.join(path, filename))
    row = [noise_level, n_train, dt]
    for i,el in enumerate(columns):
        if i<3:
            df[el][j] = row[i]
        else:
            df[el][j] = file[el][:]
```

## Number of Terms for a certain Threshold

Si vede come la "sparsity" di SINDy e la corretta individuazione del modello sia mantenuta maggiormente quando i punti sono meno soggetti a rumore. Questo si vede specialmente quando il treshold è basso, mentre per un treshold maggiore vengono mantenuti solo i parametri più rilevanti e quindi c'è minor possibilità di aggiungere termini.
Si vede però come il threshold ottimale per una corretta individuazione dei parametri ad errore zero sia 0.01, threshold nel quale i parametri iniziano già a divergere nel numero di termini per alte quantità di noise

```python
def find_len(x):
    eq_len = []
    for group in x:
        list_of_l = []
        for equation in group:
            equation = np.array2string(equation)
            s_tot = []
            for j,s in enumerate(equation.split()):
                if j==0: s = s[1:]
                if j==len(equation.split())-1 : s = s[:-1]
                try:
                    s = float((str(s)[2:]))
                    if s not in [0.0,1.0,2.0,3.0,4.0]: s_tot.append(s)
                except:
                    pass
            l = len(s_tot)
            list_of_l.append(l)    
        eq_len.append(list_of_l)
    len_array = np.array(eq_len)
    return len_array
```

```python
xaxes = [5,4,3,2,1,0]
n_trains = np.flip(np.sort(pd.unique(df['n_train'])))
n_trains = [15000,1500]

fig, AX = plt.subplots(1,2, figsize=(9.6*2,6))

fig.suptitle('Number of terms predicted by SINDy for different noise levels', size=22)#, y=0.92)

for i,ax in enumerate(AX.flatten()):
    if i<5:
        array = df[df['n_train']==n_trains[i]].sort_values('noise_level').reset_index(drop=True)['eq']
        lengths= find_len(array)
        #ax.set_yscale('log')

        def format_fn(tick_val, tick_pos):
            if int(tick_val<6):
                return th_list[int(tick_val)]
            else:
                return ''
        ax.xaxis.set_major_formatter(format_fn)

        ax.set(xlabel='Threshold value', ylabel='Number of terms', title='Training set with {} number of elements'.format(n_trains[i]))
        ax.hlines(6,xmin=0, xmax=5, colors='green', linestyles='dashed', label='True Value')

        lab = df[df['n_train']==24000].sort_values('noise_level').reset_index()['noise_level']

        for j,data in enumerate(lengths):
            ax.plot(xaxes,data, colors[j],marker='o', label=lab[j])
        ax.legend(title='Noise Level')


plt.savefig('Data/Pictures/N_of_terms_double_well.png')
```

## Parameters evolution

```python
def find_par(x,coeff, coeff2='', Just_Im = False, Just_Re = False):
    pars = []
    for a,group in enumerate(x):
        list_of_par = []
        for equations in group:
            for im_re,equation in enumerate(equations):

                if   Just_Re == True and  im_re==0: continue
                elif Just_Im == True and  im_re==1: continue
                elif Just_Im == True and Just_Re == True: break

                equation = np.array(equation)
                equation = np.array2string(equation)

                s_tot = []
                s_old = ''
                s_old_old = ''
                size=len(equation.split())
                check = False
                for j,s in enumerate(equation.split()):
                    if '\'' in s: s = s.replace('\'', '')
                    if 'b'  in s: s = s.replace('b','')
                    if s == coeff and coeff2=='':
                        try:
                            s_write = float(str(s_old)) 
                            if s_write not in ['x',2.0,3.0,4.0,5.0]: s_tot.append(s_write)
                            #if s_write not in ['x',2.0,3.0,4.0,5.0]: print('accepted',a,j,'\t',s_old,'\t',s)
                        except: pass
                    elif s == coeff2 and s_old==coeff:
                        try:
                            s_write = float(str(s_old_old)) 
                            if s_write not in ['x',2.0,3.0,4.0,5.0]: s_tot.append(s_write)
                            #if s_write not in ['x',2.0,3.0,4.0,5.0]: print('accepted',a,j,'\t',s_old_old,'\t',s_old,'\t',s)
                        except: pass
                    else: pass
                    s_old_old = str(s_old)
                    s_old = str(s)
                if s_tot != []: list_of_par.append(s_tot)
                else : list_of_par.append([0])
        pars.append(np.array(list_of_par).flatten())
    pars_array = np.array(pars)
    return pars_array
```
